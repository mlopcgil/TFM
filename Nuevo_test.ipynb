```python


import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

client_id = ''
client_secret = ''

# Autenticaci√≥n con Spotify
client_credentials_manager = SpotifyClientCredentials(client_id=client_id, client_secret=client_secret)
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)

df['track_id'] = df['uri'].apply(lambda x: x.split(':')[-1])

for i in range(len(canciones['uris'])):

  def obtener_caracteristicas(id_cancion):

      if id_cancion in df['track_id'].values:
        print("El valor ya existe en el dataset.")
      else:
        print("El valor no existe en el dataset.")
        try:
          track_info = sp.audio_features(id_cancion)
          print(track_info)

          return track_info


        except spotipy.SpotifyException as e:
            print(f"Error de Spotify API: {e}")
            return None


  caracteristicas = obtener_caracteristicas(canciones['uris'][i])

  if caracteristicas is not None:
    new_data = {
        'acousticness': caracteristicas[0]['acousticness'],
        'danceability': caracteristicas[0]['danceability'],
        'energy': caracteristicas[0]['energy'],
        'instrumentalness': caracteristicas[0]['instrumentalness'],
        'loudness': caracteristicas[0]['loudness'],
        'speechiness': caracteristicas[0]['speechiness'],
        'valence': caracteristicas[0]['valence'],
        'labels': canciones['etiquetas'][i]
    }

    new_test.loc[len(new_test)] = new_data

display(new_test)


```
